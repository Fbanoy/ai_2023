```python
def get_frames(video_name):
  cap = cv2.VideoCapture(video_name)
  images = []
  success,image = cap.read()
  images.append(image)
  count = 0
  while success:     
    success,image = cap.read()
    images.append(image)
    count += 1
  return images
```

```python
images = get_frames(video_name)
```

```python
from google.colab.patches import cv2_imshow
def imshow(image, scale=1):
  """
  Function to display an image in a specified scale
  """
  (width, length) = image.shape[:2] # Unpack width and length dimensions
  (width, length) = (int(width*scale), int(length*scale)) # Rescale
  cv2_imshow(cv2.resize(image, dsize=(length, width))) # Show resulting image
```
